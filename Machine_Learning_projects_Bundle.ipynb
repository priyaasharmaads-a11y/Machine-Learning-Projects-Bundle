# **1) EDA with univariate Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

*we will be doing univariate analysis on few variables eitehr categorical or numerical*

1) Categorical Data

*   Countplot on survived variable(give frequency for each category in the categorical col)



```python
sns.countplot(x='Survived',data=df,hue='Survived',palette='pastel',legend=False)
plt.title("Passenger Count by survival")
plt.show()
```

```python
sns.countplot(x='Pclass',data=df,hue='Pclass',palette='pastel',legend=False)
plt.title("Passenger Count by Pclass")
plt.show()
```

```python
sns.countplot(x='Embarked',data=df,hue='Embarked',palette='pastel',legend=False)
plt.title("Passenger Count by Embarkation")
plt.show()
```



*   PieChart
(for getting values in %)


```python
survived_counts = df['Survived'].value_counts()
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Survival Distribution")
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Pclass Distribution")
plt.show()
```

**Feature Engineering**
A) Feature Scaling - STANDARDIZATION

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head(10)
```

```python
df=df.iloc[:,2:]
```

```python
df.sample(5)
```

Train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),df['Purchased'],test_size=0.3,random_state=0)
X_train.shape, X_test.shape
```

StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled #getting numpy arrays here!we need to convert them
```

```python
#converting to pandas df
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)#mean is not zero and sd value is too high not standard
```

```python
np.round(X_train_scaled.describe(), 1) #ean became zero and sd is standardized
```
