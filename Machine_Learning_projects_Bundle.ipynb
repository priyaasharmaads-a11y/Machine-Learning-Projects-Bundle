# **1) EDA with univariate Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

*we will be doing univariate analysis on few variables eitehr categorical or numerical*

1) Categorical Data

*   Countplot on survived variable(give frequency for each category in the categorical col)



```python
sns.countplot(x='Survived',data=df,hue='Survived',palette='pastel',legend=False)
plt.title("Passenger Count by survival")
plt.show()
```

```python
sns.countplot(x='Pclass',data=df,hue='Pclass',palette='pastel',legend=False)
plt.title("Passenger Count by Pclass")
plt.show()
```

```python
sns.countplot(x='Embarked',data=df,hue='Embarked',palette='pastel',legend=False)
plt.title("Passenger Count by Embarkation")
plt.show()
```



*   PieChart
(for getting values in %)


```python
survived_counts = df['Survived'].value_counts()
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Survival Distribution")
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Pclass Distribution")
plt.show()
```

2. Numerical Data


*   Histogram - creating bins and visualizing the distribution




```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=10,color='skyblue',edgecolor='black')
```



*   displot (almost a histogram)
We get KDE kernel density  estimation in a curve form herein the urve touches the top part of histogram. it shows pdf - probability density function, it basically gives probabilty for that specific function.Also shows skewness of data



```python
sns.distplot(df['Age'])
```



*   Box Plot -  gives 5 measure summary, in detail in statsused for detecting outliers



```python
sns.boxplot(x='Pclass',y='Age',data=df,color='skyblue')
```

# **2) EDA with multivariate Analysis**

```python
#performing analysis on more than one column
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

1. Scatter Plot (numerical - numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='smoker',size='size',color='pink')
plt.show()
#here we did analysis on multipla variables
```

2. Bar Plot(Numerical- Categorical)

```python
sns.barplot(x='Pclass',y='Fare',data=titanic,hue='Sex',color = 'pink')
plt.show()
```

3. BOx Plot (numerical - categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue = 'Survived',color='pink')
plt.show()
#
```

4.Displot


```python
from numpy import histogram
#pdf of people who did not survive
sns.displot(data=titanic,x='Age',hue='Survived',kind='kde',fill=True)
```

In the above graph we can see that propbabilty of surviving is high for kids and adults(avg) and probability of surviving of old age people is slightly low.

5. Heatmap(categorical  - categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
#
```

6.Clustermap

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']),annot=True) #dendograms - hierarchical relation
#
```

7.Pairplot #identifies all numerical columns and pots scatter plot of every combination

```python
sns.pairplot(iris,hue='species') # to check which columns have any useful relation and which can used further.To get a bird eye view of our data
```

8. Lineplot (numerical - Numerical)
used when x axis has any time related column

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(x='year',y='passengers',data=new)
```

```python
flights.pivot_table(values='passengers',index='month',columns='year',aggfunc='sum')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year',aggfunc='sum'))
#
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year',aggfunc='sum'))
```
